```python
import pandas as pd
```

```python
# Load the dataset from your project folder
coffee_sales_df = pd.read_excel('../data/coffee_sales.xlsx')
coffee_sales_df.head()
```

```python
# check missing values
coffee_sales_df.isna().sum()
```

```python
# check duplicated rows
coffee_sales_df.duplicated().any()
```

```python
# Combine date and time into a single datetime column and drop the original date and time columns
coffee_sales_df['transaction_datetime'] = pd.to_datetime(coffee_sales_df['transaction_date'].astype(str) + ' ' + coffee_sales_df['transaction_time'].astype(str))
coffee_sales_df.drop(['transaction_date', 'transaction_time'], axis=1, inplace=True)

# Check for duplicates
duplicates = coffee_sales_df.duplicated().sum()

# Check for missing values
missing_values = coffee_sales_df.isnull().sum()

# Summary of the cleaning steps needed
cleaning_summary = {
    "Duplicates": duplicates,
    "Missing Values": missing_values
}

cleaning_summary
```

```python
# Save the preprocessed 5000 rows into a new Excel file
file_name = 'updated_coffee_sales'
coffee_sales_df.to_excel(f'../data/{file_name}.xlsx', index=False)

print(f'The new file has been saved as {file_name} in the data directory.')
```
