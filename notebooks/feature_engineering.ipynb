```python
# Ensure necessary libraries are imported
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_excel('../data/Coffee Shop Sales.xlsx')
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Assuming 'sales' is calculated as 'transaction_qty' * 'unit_price'
df['sales'] = df['transaction_qty'] * df['unit_price']

# 1. Time-Based Features
df['day_of_week'] = df['transaction_date'].dt.dayofweek  # Monday=0, Sunday=6
df['month'] = df['transaction_date'].dt.month

# Aggregate daily sales
daily_sales = df.groupby('transaction_date')['sales'].sum().reset_index()

# Creating a simple 1-day lagged sales feature
daily_sales['sales_lagged_1'] = daily_sales['sales'].shift(1)

# Note: The actual integration of lagged sales back into your original dataframe depends on your specific modeling needs.
# For predictive modeling, you might use aggregated data like 'daily_sales' directly, or you may need to merge it back to your transaction-level data depending on the granularity of your predictions.

# Ensure you replace the placeholder path with the actual path to your dataset.

daily_sales.head()  # to display the first few rows of the daily_sales DataFrame

df.head()  # to display the first few rows of the df DataFrame after feature engineering

# Merge the lagged sales feature back into the original dataframe using transaction_date as the key
df = df.merge(daily_sales[['transaction_date', 'sales_lagged_1']], on='transaction_date', how='left')

# Now that df has the 'sales_lagged_1' feature, we can display it
df[['day_of_week', 'month', 'sales', 'sales_lagged_1']].head()



```
